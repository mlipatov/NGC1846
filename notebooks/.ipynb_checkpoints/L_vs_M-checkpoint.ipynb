### log(L) versus log(M)

```python
# piecewise linear fit
def flin(x, a, y1, m1, m2):
    y2 = y1 + a * (m1 - m2)
    mask = (x < a)
    out = np.empty_like(x)
    out[mask] = m1 * x[mask] + y1
    out[~mask] = m2 * x[~mask] + y2
    return out

def xy(z):
    m = (st.logZm == z)
    x = np.log10(st.Mini[m])
    y = st.logL[m]
    p = curve_fit(flin, x, y, p0=[-0.3, -0.7, 2.2, 5.1])[0]
    xfit = np.linspace(x.min(), x.max(), 100)
    yfit = flin(xfit, *p)
    return [x, y, xfit, yfit]

x75, y75, xfit75, yfit75 = xy(-0.75)
x45, y45, xfit45, yfit45 = xy(-0.45)
x00, y00, xfit00, yfit00 = xy(0.0)
```

```python
plt.scatter(x75, y75, s=1, c='cyan', label=r'$[M/H]_{MESA}=' + str('-0.75') + '$') 
plt.plot(xfit75, yfit75, c='green')
plt.scatter(x45, y45, s=1, label=r'$[M/H]_{MESA}=' + str('-0.45') + '$') 
plt.plot(xfit45, yfit45, c='k')
plt.scatter(x00, y00, s=1, c='orange', label=r'$[M/H]_{MESA}=' + str('0.00') + '$') 
plt.plot(xfit00, yfit00, c='r')
plt.legend(markerscale=4.0)

plt.xlabel(r'$\log_{10}{\left(M_{ini} / M_\odot\right)}$')
plt.ylabel(r'$\log_{10}{\left(L / L_\odot\right)}$')
textstr = r'$log_{10}{t}: ' + str(t[0])[:5] + ' - ' + str(t[-1])[:5] + '$'
plt.text(-0.5, -3, textstr)

plt.show()
```

```python
print('Loading PARS...', end='')
with open('../data/pars_grid_2.pkl', 'rb') as f: pars = pickle.load(f)
print('Done.')
mu.Grid.std = cf.std # standard deviations of observables
mu.Grid.modulus = cf.modulus # distance modulus of the cluster
mu.Grid.pars = pars # PARS grid
```

### magnitude versus log(M) for non-rotating models

```python
for z1 in Z:
    st1 = st.copy()
    st1.select(st1.omega0 == 0)
    st1.select(st1.logZm == z1)
    st1.Z = z1
    Mini = np.logspace(np.log10(st1.Mini.min()), np.log10(st1.Mini.max()), 1000, base=10)
    mag = mu.grid(np.array([1.]), Mini, st1, pars, cf.A_V, cf.modulus)
    x = - 2.5 * np.log10(Mini)
    y = mag[..., 0, 1]
    m = ~np.isnan(y)
    print(z1, '\t', np.polyfit(x[m], y[m], 1))
    plt.scatter(x, y, s=1) 
plt.xlabel(r'$-2.5\, \log_{10}{\left(M_{ini} / M_\odot\right)}$')
plt.ylabel('F555W')
plt.gca().invert_yaxis()
plt.gca().invert_xaxis()
plt.show()
```

### magnitude versus log(M) for rotating models

```python
t1 = t[8]
z1 = -0.45
st1 = st.copy()
st1.select(st1.logZm == z1)
st1.select(st1.t == t1)
st1.Z = z1
st1.age = t1
Mini = np.logspace(np.log10(st1.Mini.min()), np.log10(st1.Mini.max()), 100, base=10)
grid = mu.Grid(st1, Mini, np.unique(st1.omega0), np.linspace(0, np.pi/2, 5), cf.A_V, verbose=True)
```

```python
x = - 2.5 * np.log10(np.repeat(grid.Mini, grid.omega0.shape[0] * grid.inc.shape[0]))
y = grid.obs[..., 0]
m = ~np.isnan(y.flatten())
p = np.polyfit(x[m], y.flatten()[m], 10, full=False) #, cov=True)
print(p)

plt.plot(x[m], np.polyval(p, x[m]), c='gray')
plt.scatter(x, y, s=1) 
plt.gca().invert_yaxis()
plt.gca().invert_xaxis()
plt.xlabel(r'$-2.5\, \log_{10}{\left(M_{ini} / M_\odot\right)}$')
plt.ylabel('F555W')
plt.show()

plt.plot(x[m], np.polyval(np.polyder(p), x[m]))
plt.gca().invert_xaxis()
plt.xlabel(r'$x \equiv -2.5\, \log_{10}{\left(M_{ini} / M_\odot\right)}$')
plt.ylabel(r'd\,F555W / d\,x')
plt.show()
```
